```python
n = 1

def table_empty(frame):
    return True if outcomes.empty else False

race_results_2019 = pd.DataFrame()

while True:
    response = requests.post(urlbase_2019 + str(n))
    soup = BeautifulSoup(response.content, "html.parser")
    results = soup.find(attrs={"class": "race-search-main"}) \
        .find("table")
    raceframes = pd.read_html(str(results))
    outcomes = raceframes[0]
    if table_empty(outcomes):
        break
    race_results_2019 = pd.concat([race_results_2019, outcomes], ignore_index=True)
    n += 1
    
race_results_2019
```
